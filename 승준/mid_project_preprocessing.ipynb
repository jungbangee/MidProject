# 컬럼 정의서
- PBLPRFR_NM : 공연명
- PBLPRFR_PLACE_NM : 공연장소명
- PBLPRFR_DE : 공연일자
- PBLPRFR_TIME : 공연시간
- GENRE_NM : 장르명
- PRC_CN : 가격
- HMPG_URL : 홈페이지 URL (삭제예정)
- SETLE_DE : 결제일자
- SETLE_TIME : 결제시간
- DSCNT_CN : 할인내용
- SETLE_PRICE : 결제금액
- MBER_GRAD_NM : 회원등급명
- BN_VALUE : 나이
- SEXDSTN_VALUE : 성별
- OC_AREA_NM : 거주지역
- RFN_AT : 환불여부

```python
# prompt: RFN_full_invert_df에서 BN_VALUE(나이) 에서 숫자만 뽑아서 int로 변환
RFN_full_invert_df = RFN_invert_df.copy()
RFN_full_invert_df['BN_VALUE'] = RFN_full_invert_df['BN_VALUE'].str.extract('(\d+)').astype(float).fillna(0).astype(int)
```

```python
RFN_full_invert_df['SEXDSTN_VALUE'] = RFN_full_invert_df['SEXDSTN_VALUE'].replace({'M': 0, 'F': 1})
RFN_full_invert_df.head()
```
