```python
import pandas as pd

DATA_PATH = "data/"
DATA_FILE = "mid-project-data.csv"
df = pd.read_csv(f"{DATA_FILE}")
#df = pd.read_csv(f"{DATA_PATH}SAC_ADAVANTK_DATA_202312.csv")

features = pd.read_csv(f"{DATA_PATH}numberofseats.csv")
features.head()


```

```python
df = pd.merge(df, features, how="left", on="PBLPRFR_PLACE_NM")
df.head()
```

```python
df['총판매석'] = df['총판매석'].astype('Int64')
df['일반판매석'] = df['일반판매석'].astype('Int64')
df['장애인석'] = df['장애인석'].astype('Int64')
df['OP석'] = df['OP석'].astype('Int64')
df['판매유보석'] = df['판매유보석'].astype('Int64')
df.head()
```

```python
def not_cancel(x):
    if x["RFN_AT"] == 'N':
        val = x["SETLE_PRICE"]
    else :
        val = 0
    return val

df["취소되지않은결제액"] = df.apply(not_cancel, axis = 1)
```

```python
cols = ["PBLPRFR_NM", "PBLPRFR_PLACE_NM", "PBLPRFR_DE", "PBLPRFR_TIME"]
agg_dict = {
    "GENRE_NM": "count",
    "SETLE_PRICE": "sum",
    "취소되지않은결제액": "sum",
    "RFN_AT": lambda x : (x == 'N').sum(),
    "총판매석": lambda x : x.unique()[0]
    
}


df['취소되지않은결제액'] = df['취소되지않은결제액'].astype('Int64')
df['SETLE_PRICE'] = df['SETLE_PRICE'].astype('Int64')
new_df = df.groupby(cols).agg(agg_dict)

#new_df.to_csv("tmp.csv")
```
