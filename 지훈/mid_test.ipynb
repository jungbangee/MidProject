```python
import pandas as pd
df = pd.read_csv(f"{DATA_PATH}SAC_ADAVANTK_DATA_202312.csv")
df
```

```python
# 한글화
df.rename(columns ={"PBLPRFR_NM" : "공연명","PBLPRFR_PLACE_NM" : "공연장소명", "PBLPRFR_DE" : "공연일자", "PBLPRFR_TIME" : "공연시간", "GENRE_NM" : "장르명", "PRC_CN" : "가격내용"
, "HMPG_URL" : "홈페이지URL", "SETLE_DE" : "결제일자", "SETLE_TIME" : "결제시간", "DSCNT_CN" : "할인내용", "SETLE_PRICE" : "결제금액", "MBER_GRAD_NM" : "회원등급명", "BN_VALUE" : "나이값"
, "SEXDSTN_VALUE" : "성별값", "OC_AREA_NM" : "거주지역명", "RFN_AT" : "환불여부"}, inplace=True)
df
```

```python
# 1. 성별은 M → 0 / F → 1 로 변환
# 2. RFN_AT(환불여부)는 N → 0 / Y → 1 로 변환
df = df.replace({'성별값': {'M':'0', 'F' : '1'}, '환불여부' : {'N':'0', 'Y':'1'}})
df
```

```python
# 성별에 따른 공현 선호도(M → 0 / F → 1)
선호도_성별 = df.groupby(["공연명", "성별값"])["결제금액"].sum().reset_index()
선호도_성별.sort_values(by=['공연명',"결제금액"], ascending=[True,False]).reset_index()
```

```python
# 장르별 공연 선호도
선호도_장르별 = df.groupby("장르명")["결제금액"].sum().reset_index()
선호도_장르별.sort_values(by="결제금액", ascending=False)
```

```python
# 나이별 공연 선호도(20위까지)
선호도_나이대별 = df.groupby(["나이값","성별값","장르명"])["결제금액"].sum().reset_index()
선호도_나이대별.sort_values(by="결제금액", ascending=False).head(20)
```

```python
# 나이대, 성별, 장르별 결제 금액 합계 및 카운트 계산
선호도_나이대별 = df.groupby(['나이값', '성별값', '장르명']).agg(
    결제금액=('결제금액', 'sum'),
    카운트=('결제금액', 'size')  # 카운트 추가
).reset_index()

# 결제 금액 기준으로 정렬
선호도_나이대별 = 선호도_나이대별.sort_values(by='결제금액', ascending=False)

# 상위 20개 결과만 보기
선호도_나이대별.head(20)
```

```python
# 나이대 및 성별별 결제 금액 합계 계산
선호도_나이대별_성별 = df.groupby(['나이값', '성별값', '장르명'])['결제금액'].sum().reset_index()

# 전체 결제 금액을 나이대별로 계산
전체_결제금액 = 선호도_나이대별_성별.groupby(['나이값', '장르명'])['결제금액'].sum().reset_index()
전체_결제금액.rename(columns={'결제금액': '전체_결제금액'}, inplace=True)

# 나이대별 성별 비율 계산
선호도_나이대별_비율 = pd.merge(선호도_나이대별_성별, 전체_결제금액, on=['나이값', '장르명'])
선호도_나이대별_비율['비율'] = 선호도_나이대별_비율['결제금액'] / 선호도_나이대별_비율['전체_결제금액'] * 100

# 결과 정렬 및 출력
선호도_나이대별_비율.sort_values(by=['나이값', '장르명', '성별값'], ascending=True, inplace=True)
선호도_나이대별_비율
```

```python
# 나이대, 성별, 장르별 결제 금액 합계 및 카운트 계산
선호도_나이대별_성별_장르 = df.groupby(['나이값', '성별값', '장르명']).agg(
    결제금액=('결제금액', 'sum'),
    카운트=('결제금액', 'size')
).reset_index()

# 전체 결제 금액을 나이대 및 장르별로 계산
전체_결제금액 = 선호도_나이대별_성별_장르.groupby(['나이값', '장르명'])['결제금액'].sum().reset_index()
전체_결제금액.rename(columns={'결제금액': '전체_결제금액'}, inplace=True)

# 나이대, 성별, 장르별 비율 계산
선호도_나이대별_비율 = pd.merge(선호도_나이대별_성별_장르, 전체_결제금액, on=['나이값', '장르명'])
선호도_나이대별_비율['비율'] = 선호도_나이대별_비율['결제금액'] / 선호도_나이대별_비율['전체_결제금액'] * 100

# 결과 정렬 및 출력
선호도_나이대별_비율.sort_values(by=['나이값', '장르명', '성별값'], ascending=True, inplace=True)
선호도_나이대별_비율.head(20)
```

```python
# 선호도 20대
선호도_20대 = 선호도_나이대별_비율[선호도_나이대별_비율['나이값'] == '20대']
선호도_20대
```
